```python
#整合divvy_trips数据
import pandas as pd
#读取文件名
import os
from dateutil.parser import parse
def trans_time(f):
    f1=f.apply(lambda x:parse(x))
    f1=f1.apply(lambda x:str(x.year)+'-'+'{:02}'.format(x.month))
    return f1
filelist=os.listdir('D:\\python-document\\divvy_trips')
a='D:\\python-document\\divvy_trips\\'
#huizongz=pd.DataFrame(columns=['attraction_id', 'review_time','from_is_sub', 'from_count_id','from_tripduration','to_is_sub', 'to_count_id','to_tripduration','from_dpcapacity','to_dpcapacity'])
#huizong1=pd.DataFrame(columns=['attraction_id', 'review_time','is_sub','to_count_id','tripduration'])
meidi_to=pd.DataFrame(columns=['trip_time','to_station_id','to_is_sub','to_count_id','to_tripduration'])
meidi_from=pd.DataFrame(columns=['trip_time','from_station_id','from_is_sub','from_count_id','from_tripduration'])
for item in filelist:
    medi=pd.read_csv(a+item)
    medi.columns=['trip_id', 'starttime', 'stoptime', 'bikeid', 'tripduration',
       'from_station_id', 'from_station_name', 'to_station_id',
       'to_station_name', 'usertype', 'gender', 'birthday']
    medi['trip_time']=medi['stoptime']
    medi['trip_time']=trans_time(medi['trip_time'])
    medi['count_id']=1
    medi['is_sub']=medi['usertype'].apply(lambda x:1 if x=='Subscriber' else 0)
    medi_group=medi.groupby(['trip_time','to_station_id'])['is_sub','count_id','tripduration'].sum().reset_index()
    medi_group.columns=['trip_time','to_station_id','to_is_sub','to_count_id','to_tripduration']
    medi_group_1=medi.groupby(['trip_time','from_station_id'])['is_sub','count_id','tripduration'].sum().reset_index()
    medi_group_1.columns=['trip_time','from_station_id','from_is_sub','from_count_id','from_tripduration']
    meidi_to=pd.concat([meidi_to,medi_group])
    meidi_from=pd.concat([meidi_from,medi_group_1])
    print(item,':完成')
meidi_to.to_csv('D:\\python-document\\divvy_trips\\to_station.csv')
meidi_from.to_csv('D:\\python-document\\divvy_trips\\from_station.csv')
    
```
